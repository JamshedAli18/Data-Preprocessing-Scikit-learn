```python
import pandas as pd
```

```python
df = pd.read_csv('/content/cars.csv')
```

```python
df.head()
```

```python
brands_count = df['brand'].value_counts()
```

```python
brands_count
```

```python
brands_to_replace = brands_count[brands_count<100].index
```

```python
brands_to_replace
```

```python
df['brand'] = df['brand'].replace(brands_to_replace,'other')
```

```python
df.head()
```

```python
df['brand'].value_counts()
```

```python
columns_to_encode = ['brand', 'fuel', 'owner']
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse=False, drop='first')
```

```python
encoded_data = ohe.fit_transform(df[columns_to_encode])
```

```python
encoded_data
```

```python
columns_names = ohe.get_feature_names_out(columns_to_encode)
```

```python
columns_names
```

```python
encoded_data_df = pd.DataFrame(encoded_data, columns=columns_names)
```

```python
encoded_data_df.head()
```

```python
encoded_df = pd.concat([df.drop(columns_to_encode,axis=1),encoded_data_df],axis=1)
```
