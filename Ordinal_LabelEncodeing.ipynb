```python
import pandas as pd

# Sample data
data = {
    'age': [25, 30, 22, 28, 35],
    'gender': ['male', 'female', 'female', 'male', 'female'],
    'review': ['good', 'average', 'good', 'excellent', 'good'],
    'education': ['Bachelor', 'Master', 'Bachelor', 'PhD', 'Master'],
    'purchase': [True, False, True, True, False]
}

# Create DataFrame
df = pd.DataFrame(data)
```

```python
df
```

```python
df = df.drop(['age','gender'],axis=1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train,y_test = train_test_split(df[['review','education']],df['purchase'],test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
```

```python
X_train
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
y_train = encoder.fit_transform(y_train)
```

```python
y_train
```

```python
df['purchase'] = encoder.fit_transform(df[['purchase']])
```

```python
df
```

```python
df[['review','education']] = oe.fit_transform(df[['review','education']])
```
