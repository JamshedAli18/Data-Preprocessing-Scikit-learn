```python
import pandas as pd
data = {
   'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 32, 37, 45, 52],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Department': ['HR', 'Engineering', 'Marketing', 'Finance', 'Engineering'],
    'City': ['New York', 'San Francisco', 'Chicago', 'Austin', 'New York'],
    'Education': ['Bachelor', 'Master', 'PhD', 'Bachelor', 'Master'],
    'Performance': ['Excellent', 'Good', 'Average', 'Good', 'Excellent']
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
x = df.drop('Performance',axis=1)
y = df['Performance']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder(categories=[['Bachelor', 'Master', 'PhD']])
```

```python
df['Education'] = encoder.fit_transform(df[['Education']])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelEncoder = LabelEncoder()
```

```python
df['Performance'] = labelEncoder.fit_transform(df[['Performance']])
```
