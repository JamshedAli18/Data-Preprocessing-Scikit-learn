```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('/content/covid_toy.csv')
```

```python
df.head()
```

```python
df.isnull().any()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transfromer = ColumnTransformer(transformers=[
    ('trf1',SimpleImputer(),['fever']),
    ('trf2',OneHotEncoder(sparse=False,drop='first'),['gender','city']),
    ('trf3',OrdinalEncoder(categories=[['Mild','Strong']]),['cough'])
],remainder='passthrough')
```

```python
x = transfromer.fit_transform(df)
```

```python
df1 = pd.DataFrame(x)
```

```python
df1
```

```python
transformed_data = transfromer.fit_transform(df)

# Get feature names
feature_names_fever_cough = ['fever', 'cough']
feature_names_gender_city = transfromer.named_transformers_['trf2'].get_feature_names_out(['gender', 'city'])
feature_names_remainder = ['age', 'has_covid']

# Combine all feature names
all_feature_names = feature_names_fever_cough + list(feature_names_gender_city) + feature_names_remainder

print(all_feature_names)
```

```python
transformed_df = pd.DataFrame(transformed_data, columns=all_feature_names)
```
